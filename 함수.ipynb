```python
my_dict = {"10/26" : [100, 130, 100, 100],
           "10/27" : [10, 12, 10, 11]}

def print_value_by_key(dic,key):
    print(dic[key])
    print(dic[key][2]) #요소 하나 인덱스로 가져오기
print_value_by_key(my_dict, "10/26")
```

```python
my_dict = {"10/26" : [100, 130, 100, 100],
           "10/27" : [10, 12, 10, 11]}
def safe_print_value(dic, key):
    #값이 없을 경우 "데이터가 없습니다"출력
    result = dic.get(key,'데이터가 없습니다.')
    print(result)
safe_print_value(my_dict,"10/26")
safe_print_value(my_dict,"10/28")
```

```python

def make_list(string):
    u_list = []
    for i in string:
        u_list.append(i)
    return u_list

result = make_list("abcd")
print(result)
```
