```python
exchange  = {'달러':1167,'엔':1096,'유로':1268,'위안':171}
u_cash = input()
new_cash = u_cash.split(" ")
print(new_cash)
#입력받은 값에서 숫자만 뺀 '금액' 정수화
money = int(new_cash[0])
#환율 
doller = exchange['달러']
yen = exchange['엔']
euro = exchange['유로']
yuan = exchange['위안']

if new_cash[1] == '달러':
    result = money*doller
    print(f"{result:.2f}원")
elif new_cash[1] == '엔':
    result = money*yen
    print(f"{result:.2f}원")
elif new_cash[1] == '유로':
    result = money*euro
    print(f"{result:.2f}원")
else:
    print(f"{money*yuan:.2f} 원")
```

```python
exchange  = {'달러':1167,'엔':1096,'유로':1268,'위안':171}
u_cash = input()
new_cash = u_cash.split(" ")
print(new_cash)
money = int(new_cash[0])
currency = new_cash[1]
print(money)
print(currency)

if currency in exchange:
    result = money * exchange[currency]
    print(f"{result:.2f} 원")
else:
    print("알 수 없는 통화입니다.")
    
```

-문제126 코드 줄이기 (딕셔너리 활용 리팩토링)
============================================

```python
post = input('우편번호:')
gu = post[:3]
if gu in ['010','011','012']:
    print('강북구')
elif gu in ['013','014','015']:
    print('도봉구')
elif gu in ['016','017','018']:
    print('노원구')
else:
    print('우편번호를 확인하세요')
```

```python
post = input('우편번호:')
gu_code = post[:3]

# 우편번호 정보를 담은 딕셔너리
gu_map = {
    '010': '강북구', '011': '강북구', '012': '강북구',
    '013': '도봉구', '014': '도봉구', '015': '도봉구',
    '016': '노원구', '017': '노원구', '018': '노원구'
}

# 딕셔너리의 get 메서드를 사용하여 출력
# (찾는 값이 없으면 '우편번호를 확인하세요'를 기본값으로 출력)
print(gu_map.get(gu_code, '우편번호를 확인하세요'))
```
